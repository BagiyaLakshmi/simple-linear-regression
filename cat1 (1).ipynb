```python
import pandas as pd
import numpy as np
import math
import seaborn as sn
import operator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.rand(1000, 2) * 2
```

```python
data
```

```python
X, Y = np.split(data,2,axis=1)
```

```python
plt.plot(X,Y)
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
m = len(X)

# using the formula to calculate m & c
numer = 0
denom = 0
for i in range(m):
  numer += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print (f'm = {m} \nc = {c}')
```

```python
max_x = np.max(X) + 100
min_x = np.min(Y) - 100

# calculating line values x and y
x = np.linspace (min_x, max_x, 100)
y = c + m * x

plt.plot(x, y, color='#58b970', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='data points')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
ss_t = 0 #total sum of squares
ss_r = 0 #total sum of square of residuals

for i in range(int(120)):
  y_pred = c + m * X[i]
  ss_t += (Y[i] - mean_y) ** 2
  ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)

print(r2)
```

```python
#calculating mean absolute error
sum1=0
diff=0
for i in range(int(m)):
    y_pred = c + m * X[i]
    diff=abs(y_pred-Y[i])
    sum1+=diff
total=sum1/m
total
```

```python
#calculating rmse
sum1=0
diff1=0
for i in range(int(m)):
    y_pred = c + m * X[i]
    diff1 = (y_pred-Y[i]) ** 2
    sum1 += diff1
    
total1=math.sqrt(sum1/m)
total1
```

### Gradient descent for best fit

```python
# Building the model
m = 0
c = 0

L = 0.1  # The learning Rate
epochs = 400  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

```python
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
